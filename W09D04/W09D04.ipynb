```python
WHAT ARE API'S

API'S ARE GENERALLY USE TO EXCHANGE DATA (JSON IS MOSTLY USED TO INTERCHANGE THE DATA WITH THE APIS)

CONSIDER THAT YOU ARE IN A RESTRAUNT

1) CALL THE WAITER
2) YOU WILL TELL HIM THE ORDER DETAILS (LIKE PIZZA, COLD DRINK)
3) WAITER WILL GO BACK IN THE KITCHEN AND TELL THE CHEF TO MAKE A PIZZA AND PREPARE THE COLD DRINKS
IN GLASSES.
4) WAITER WILL COME BACK AFTER SOME TIME WITH PREPARED FOOD.


API ARE ANALOGOUS TO WAITER API === WAITER.

1) CALL THE API
2) YOU WILL GIVE IT SOME INFO(PAYLOAD IF IT IS A POST REQUEST) OR PARAMETERS(IF IT IS A GET REQUEST)
3) API WILL GO BACK TO THE SERVER SIDE LOGIC, WHICH MIGHT ACCESS THE DATABASE AND PREPARE THE DATA
4) API WILL COME BACK WITH THE RESPONSE.
```
